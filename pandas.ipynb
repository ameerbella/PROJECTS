```python
import pandas as pd
```

```python
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
print(df)
```

```python
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
print(df)
```

```python
#grouping
df['Age'] = [25, 30]
print(df.groupby('Gender')['Age'].mean())
```

```python
# Merge two DataFrames
data1 = {'Name': ['sarah', 'james'], 'Age': [25, 27]}
df1 = pd.DataFrame(data1)
data2 = {'Name':['sarah','james'],'subject': ['maths', 'english'], 'marks': [70, 77]}
df2 = pd.DataFrame(data2)
print(pd.merge(df1, df2, on='Name'))
```

```python
# Join by index
data1 = {'Name': ['sarah', 'james'], 'Age': [25, 27]}
df3 = pd.DataFrame(data1)
data2 = {'Name':['sarah','james'],'subject': ['maths', 'english'], 'marks': [70, 77]}
df4 = pd.DataFrame(data2)
df3.set_index('Name').join(df4.set_index('Name'))
```
